## Requirement 1

```python
import requests
import time
def send_request(colors):
    requests.post("https://si568.umsi.run/change?key=eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJ1c2VyIjoiaHVhbmdndWFAdW1pY2guZWR1In0.0eGHkw5T-oPWEg_NcgBtIAATq4Pcp-BWsUs-OVdr1BY&device=TWIG", json={'values': colors})

def turn_off():
    requests.post("https://si568.umsi.run/off?key=eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJ1c2VyIjoiaHVhbmdndWFAdW1pY2guZWR1In0.0eGHkw5T-oPWEg_NcgBtIAATq4Pcp-BWsUs-OVdr1BY&device=TWIG")

def requirement1():
    color_sequence = [
        "0, 0, 255, 0", # blue
        "255, 0, 0, 0", # red
        "0, 255, 0, 0" # green
    ]
    colors = ""
    for color in color_sequence:
        for i in range(300):
            colors += color + ", "
        colors = colors[:-2]
        send_request(colors)
        print(colors)
        colors = ""
        time.sleep(1)
    turn_off()

requirement1()
```

## Requirement 2

```python
def requirement2():
    color_sequence = [
        "0, 0, 255", # blue
        "255, 0, 0", # red
        "0, 255, 0" # green
    ]
    no_color = "0, 0, 0, 0"
    no_colors = ""
    colors = ""
    for i in range(300):
        no_colors += no_color + ", "
    no_colors = no_colors[:-2]
    for color in color_sequence:
        for i in range(25):
            colors += color + ", "
        colors = colors[:-2]
        colors = no_colors + ", " + colors;
        print(colors)
        send_request(colors)
        colors = ""
        time.sleep(1)
    turn_off()

requirement2()
```

## Requirement 3

```python
color_sequence_1 = [
    "251, 236, 93, 0",
    "0, 0, 255, 0"
]
color_sequence_2 = [
    "251, 236, 93",
    "0, 0, 255"
]

def requirement3():
    colors = ""
    for j in range(300):
        if(j % 2 == 0):
            colors += color_sequence_1[0] + ", "
        else:
            colors += color_sequence_1[1] + ", "
    for j in range(256):
        if(j % 2 == 0):
            colors += color_sequence_2[0] + ", "
        else:
            colors += color_sequence_2[1] + ", "
    colors = colors[:-2]
    send_request(colors)
    print(colors)
    colors = ""
    time.sleep(10)
    turn_off()

requirement3()
```

## Requirement 4 - (a)

```python
color_sequence_1 = [
    "251, 236, 93, 0",
    "0, 0, 255, 0"
]
color_sequence_2 = [
    "251, 236, 93",
    "0, 0, 255"
]

def requirement3():
    colors = ""
    for i in range(5):
        for j in range(300):
            if((j % 2 == 0 and i % 2 == 0) or (j % 2 == 1 and i % 2 == 1)):
                colors += color_sequence_1[0] + ", "
            else:
                colors += color_sequence_1[1] + ", "
        for j in range(256):
            if((j % 2 == 0 and i % 2 == 0) or (j % 2 == 1 and i % 2 == 1)):
                colors += color_sequence_2[0] + ", "
            else:
                colors += color_sequence_2[1] + ", "
        colors = colors[:-2]
        send_request(colors)
        print(colors)
        colors = ""
        time.sleep(1)
    turn_off()

requirement3()
```

## Requirement 4 - (b)

```python
color_sequence_1 = [
    "251, 236, 93, 0",
    "0, 0, 255, 0"
]
color_sequence_2 = [
    "251, 236, 93",
    "0, 0, 255"
]

def requirement4():
    colors = ""
    for i in range(5):
        if (i % 2 == 0):
            for j in range(300):
                colors += color_sequence_1[0] + ", "
            for j in range(256):
                colors += color_sequence_2[0] + ", "
        else:
            for j in range(300):
                colors += color_sequence_1[1] + ", "
            for j in range(256):
                colors += color_sequence_2[1] + ", "

        colors = colors[:-2]
        send_request(colors)
        print(colors)
        colors = ""
        time.sleep(1)
    turn_off()

requirement4()
```

## Requirement 5 - Rainbow Snake

```python
color_sequence = "255, 0, 0, 255, 165, 0, 255, 255, 0, 0, 128, 0, 0, 0, 255, 75, 0, 130, 148, 0, 211"
no_color = "0, 0, 0, 0"
no_color_2 = "0, 0, 0"

def requirement5():
    colors = ""
    for i in range(300):
        colors += no_color + ", "
    colors += color_sequence
    for j in range(250):
        send_request(colors)
        colors = no_color_2 + ", " + colors
    turn_off()

requirement5()
        
```
